```python
image_size =50
path1 = r"C:\Users\Rohit\Desktop\Data Science Projects\Deep Learning Case Studies\DL Exam\Dataset\ANN\Pizza Refund\train"
group = ['Burnt_pizza','Good_pizza']
input_image = []
for i in group:
    folders = os.path.join(path1, i)
    label = group.index(i)
    for image in os.listdir(folders):
        image_path = os.path.join(folders , image)
        if(image_path[-3:] != "gif"):
            image_array = cv2.imread(image_path,0)
            image_array =cv2.resize(image_array,(image_size,image_size))
            input_image.append([image_array , label])
```

# 9. Hyper Parameter Tuning

```python
import keras_tuner as kt
```

```python
def build_model(hp):
    model = keras.Sequential()
    for i in range(hp.Int('num_layers', 2, 20)):
        model.add(keras.layers.Dense(hp.Choice('units', [8, 16, 32]), activation='relu'))
    model.add(keras.layers.Dense(1, activation='relu'))
    model.compile(optimizer = "adam", loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
    return model
```

```python
tuner = kt.RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=5)
```
